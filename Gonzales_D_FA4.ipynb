```python
#3

normal = [67, 70, 63, 65, 68, 60, 70, 64, 69, 61, 66, 65, 71, 62, 66, 68, 64, 67, 62, 66, 65, 63, 66, 65, 63, 69, 62, 67, 59, 66, 65, 63, 65, 60, 67, 64, 68, 61, 69, 65, 62, 67, 70, 62, 63, 68, 64, 65, 61, 66]

def moments_about_value(data, value):
    n = len(data)
    first_moment = sum([(i - value) for i in data]) / n
    second_moment = sum([(i - value) ** 2 for i in data]) / n
    third_moment = sum([(i - value) ** 3 for i in data]) / n
    fourth_moment = sum([(i - value) ** 4 for i in data]) / n
    return first_moment, second_moment, third_moment, fourth_moment

first_moment_75, second_moment_75, third_moment_75, fourth_moment_75 = moments_about_value(normal, 75)

print(f"First moment about 75 : {first_moment_75:.2f}")
print(f"Second moment about 75: {second_moment_75:.2f}")
print(f"Third moment about 75: {third_moment_75:.2f}")
print(f"Fourth moment about 75: {fourth_moment_75:.2f}")

#4

m_1 = first_moment_75
m_2 = second_moment_75
m_3 = third_moment_75
m_4 = fourth_moment_75

m2 = m_2 - (m_1 ** 2)
m3 = m_3 - (3 * m_1 * m_2) + (2 * m_1 ** 3)
m4 = m_4 - (4 * m_1 * m_3) + (6 * m_1 ** 2 * m_2) - (3 * m_1 ** 4)

print("-----------------------------------------")
print(f"Relation between second moment (m2): {m2:.2f} ")
print(f"Relation betwee third moment (m3): {m3:.2f}")
print(f"Relation between fourth moment (m4): {m4:.2f}")

# As we can see the values m(i), it is equivalent to the values that we have on -
# problem 2 for moments in Normal set (female height measurements).


```
