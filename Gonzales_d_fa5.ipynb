```python
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given population and probabilities from Problem 8.17
population = [9, 12, 15]
p = [1/3, 1/3, 1/3]

samples = list(itertools.product(population, repeat=2))

sample_means = [np.mean(sample) for sample in samples]


mean_counts = pd.Series(sample_means).value_counts(normalize=True)


sampling_dist = pd.DataFrame({
    'x̄': mean_counts.index,
    'p(x̄)': mean_counts.values
})


sampling_dist['x̄ * p(x̄)'] = sampling_dist['x̄'] * sampling_dist['p(x̄)']
sampling_dist['x̄² * p(x̄)'] = sampling_dist['x̄']**2 * sampling_dist['p(x̄)']
sampling_dist['x̄² * p(x̄)'].sum()


plt.bar(sampling_dist['x̄'], sampling_dist['p(x̄)'], color='skyblue', edgecolor='black')
plt.title('Sampling Distribution of the Mean')
plt.xlabel('Sample Mean (x̄)')
plt.ylabel('Probability (p(x̄))')
plt.grid(True)
plt.show()

mean_xbar = sampling_dist['x̄ * p(x̄)'].sum()
variance_xbar = sampling_dist['x̄² * p(x̄)'].sum() - mean_xbar**2


mu_population = 12
variance_population = 6

expected_mean = sampling_dist['x̄ * p(x̄)'].sum()
expected_variance = sampling_dist['x̄² * p(x̄)'].sum() - expected_mean**2


expected_mean, expected_variance
```

```python
import itertools
import numpy as np

population = [3, 7, 11, 15]


population_mean = np.mean(population)


population_std = np.std(population, ddof=0) 


samples = list(itertools.product(population, repeat=2))

sample_means = [np.mean(sample) for sample in samples]


sampling_mean = np.mean(sample_means)

sampling_std = np.std(sample_means, ddof=0)  
n = 2
expected_sampling_mean = population_mean
expected_sampling_std = population_std / np.sqrt(n)


population_mean, population_std, sampling_mean, sampling_std, expected_sampling_mean, expected_sampling_std
```

```python
from scipy.stats import binom


n = 200  
p = 0.5 
x_a = int(0.4 * n)  
prob_a = binom.cdf(x_a - 1, n, p)

x_b1 = int(0.43 * n)  
x_b2 = int(0.57 * n)  
prob_b = binom.cdf(x_b2, n, p) - binom.cdf(x_b1 - 1, n, p)

x_c = int(0.54 * n)  
prob_c = 1 - binom.cdf(x_c, n, p)

prob_a, prob_b, prob_c
```

```python
import itertools
import numpy as np
import pandas as pd

x_values = [6, 9, 12, 15, 18]
probabilities = [0.1, 0.2, 0.4, 0.2, 0.1]

samples = list(itertools.product(x_values, repeat=2))

means = [np.mean(sample) for sample in samples]

sample_probabilities = [probabilities[x_values.index(sample[0])] * probabilities[x_values.index(sample[1])] for sample in samples]

df = pd.DataFrame({
    'Sample': samples,
    'Mean': means,
    'Probability': sample_probabilities
})


df

```
